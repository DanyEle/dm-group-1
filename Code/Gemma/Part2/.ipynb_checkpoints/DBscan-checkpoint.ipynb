```python
import os
import sys
import math
import numpy as np
import matplotlib as mpl
mpl.use('TkAgg')
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from scipy.spatial.distance import pdist,squareform
from sklearn.cluster import DBSCAN
sys.path.insert(0, '../../Daniele')
from k_means import convert_education_to_numerical_attribute
```

```python
def Average(lst): 
    return sum(lst) / len(lst) 
```

```python
filePath = "../../../Dataset/credit_default_cleaned.csv"
df2 = pd.read_csv(filePath)
```

```python
df2.head()
```

```python
df2 = df2.drop(["sex", "status", "credit_default"], axis=1)
dataNew = convert_education_to_numerical_attribute(df2)
```

```python
scaler = MinMaxScaler()  #normalization
df1 = scaler.fit_transform(dataNew.values)  #df1 is normalized
dbscan = DBSCAN(eps=0.25, min_samples=32)
dbscan.fit(df1)
plt.scatter(df2['limit'], df2['age'], c=dbscan.labels_, s=20)
plt.tick_params(axis='both', which='major', labelsize=22)
plt.show()
```
