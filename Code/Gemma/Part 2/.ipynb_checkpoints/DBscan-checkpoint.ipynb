```python
import math
import numpy as np
import matplotlib as mpl
mpl.use('TkAgg')
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from scipy.spatial.distance import pdist,squareform
```

```python
def Average(lst): 
    return sum(lst) / len(lst) 
```

```python
filePath = "../../../Dataset/credit_default_cleaned.csv"
df2 = pd.read_csv(filePath)
```

```python
df2.head()
```

```python
df2 = df2.drop(["sex", "education", "status", "credit_default"], axis=1)
```

```python
scaler = MinMaxScaler()  #normalization
df1 = scaler.fit_transform(df2.values)  #df1 is normalized
dist = squareform(pdist(df1,metric="euclidean")) #distance matrix
mean_values=list()
for k in range(3,100):
    for d in dist:
        index_kth_distance = np.argsort(d)[k]
        kth_distances.append(d[index_kth_distance])
    mean_values.append(Average(kth_distances))
plt.plot(range(0, len(mean_values)), sorted(mean_values))
plt.show()
```
