```python
def getDistinctValues(df, column):
    print(column)
    values=df[column].unique()
    if(df[column].dtype == 'int64'):
        values.sort()
    print(values)
```

```python
columns=data.columns
for c in columns:
    getDistinctValues(data, str(c))
```

```python
def correctype(column):
    for x in column:
        if(type(x) != column.dtype):
            print(type(x), ": " ,str(x))
```

```python
for c in data.columns:
    values=data[str(c)].unique()
    print("NAME: ", c , " TYPE: ", values.dtype)
    correctype(values)
```
